```python
#load the model of sentimental pretrained model
import flair
flair_sentiment = flair.models.TextClassifier.load('en-sentiment')
```

**Annotating data for NER: CATALYST sentences**

```python
#Access the actual catalyst entities
df_catalyst = pd.read_csv(r'C:\Users\kumar\OneDrive - IIT Delhi\CPCB\PROF. HARI\data_final\catalyst\catalys_entit_final.csv')
#Delete the unwanted columns
df_catalyst = df_catalyst.drop(['Unnamed: 0', 'Unnamed: 3', '7738'], axis =1)
#Catalyst entities in list form
name_catalyst = list(df_catalyst['section name'])
```

```python
#load the data of catalyst sentences
data = pd.read_csv(r'C:\Users\kumar\OneDrive - IIT Delhi\CPCB\PROF. HARI\data_final\CLASSIFICATION\df_cat.csv')
data = data.drop_duplicates(['sentence'], keep ='first').reset_index(drop =True)
data = data.drop(['Unnamed: 0'], axis =1)
data.head()
```

```python
#Cleaning the each sentence in data frame
sentence = []
import re
for i in np.arange(0,data.shape[0], 1):
    pattern = r'\[[^()]*\]'
    t = re.sub(pattern,'', data.sentence[i])
    sentence.append(t)
data['sentence'] = sentence
#Shows first 5 rows
data.head()
```

```python
#string to be searched
string = 'catalyst'                          
add = []                               

for z in np.arange(0,data.shape[0], 1):
    #Access the each sentence
    s = data.sentence[z]
    if string in s:
        #extract chemical species using chemdataextractor
        doc = Document(s)
        ent = doc.cems                       
        comp = []
    for i in range(len(ent)):
        #convert span object to str
        comp.append(ent[i].text)           
    if ent != []:
        #sentiment analysis
        si = flair.data.Sentence(s)         
        flair_sentiment.predict(si)
        total_sentiment = si.labels
        d = total_sentiment[0].to_dict()
        #if positive label then add in csv
        if d['value'] == 'POSITIVE':         
            comp = set(comp)
        if s[-1] == '.':
            s = s[:-1]
            sl = word_tokenize(s)
            x = word_tokenize(s)
            add.append(['__','__','__','__'])
            for i in range(len(sl)):
                if sl[i] in name_catalyst:
                    sl[i] = 'catalyst'
                else:
                    sl[i] = 0
                add.append([x[i],'NN', 0, sl[i]])
```

```python
#Data store in data frame form
df = pd.DataFrame(add)
#Shows the first 5 rows
df.head()
```

```python
#Delete the first row
df =  df.iloc[1: , :]
#Reset the index and drop previous
df = df.reset_index(drop = True)
#Shows the first 5 rows
df.head()
```

```python
#Save the data frame in text format
np.savetxt(r'/home/user3/Documents/avan_phd/Objective_nlp/data_file/df_annotated_ner_cat.txt', df.values, fmt='%s')
```

```python
#Delete the duplicate entries
data_pro = data_pro.drop_duplicates(['sentence'], keep ='first').reset_index(drop = True)
```

```python
#Cleaning the each sentence in data frame
sentence = []
import re
for i in np.arange(0,data_pro.shape[0], 1):
    pattern = r'\[[^()]*\]'
    t = re.sub(pattern,'', data_pro.sentence[i])
    sentence.append(t)
data_pro['sentence'] = sentence
#Shows first 5 rows
data_pro.head()
```

```python
#List of lists utilized
all_string = []
name_process = []
sentence_list = []
add1 = []
#Custom dictionary of process parameters
process = ['temperature', 'pressure', 'weight', 'concentration', 'molar ratio',
           'molarity','Molality','gram','atm','celcius', 'kelvin' , 'volume', 'porocity']

replacements = {'K': 'kelvin', 'atm': 'pressure',
               '°C': 'celcius','bar': 'pressure',
                'g' : 'gram','m': 'molarity',
                'M': 'Molality','Mpa': 'pressure',
                'pa' : 'pressure', }
#Help in dealing with units of parameters
replacer = replacements.get

for i in np.arange(0,data_pro.shape[0],1):
    s = data_pro.sentence[i]
    sentence_list = nltk.word_tokenize(s)
    sentence_list = [replacer(n, n) for n in sentence_list]
    sentence_list = " ".join(sentence_list)
    for elem in process:
        if elem in sentence_list:
            #sentiment analysis
            si = flair.data.Sentence(sentence_list)         
            flair_sentiment.predict(si)
            total_sentiment = si.labels
            d = total_sentiment[0].to_dict()
            if d['value'] == 'POSITIVE':
                if s[-1] == '.':
                    s = s[:-1]
                    sl = word_tokenize(sentence_list)
                    x = word_tokenize(sentence_list)
                    add1.append(['__','__','__','__'])
                    for i in range(len(sl)):
                        if sl[i] in process:
                            sl[i] = 'process'
                        else:
                            sl[i] = 0
                        add1.append([x[i],'NN', 0, sl[i]])   
```

```python
#Stored in data frame 
df = pd.DataFrame(add1)
#Delete the first row
df =  df.iloc[1: , :]
#Reset the index and drop previous
df = df.reset_index(drop = True)
#Shows the first 5 rows
df.head()
```

**Annotating data for NER: PROCESS-CATALYST sentences**

```python
#load the data of catalyst and process
data_cat_pro = pd.read_csv(r'C:\Users\kumar\OneDrive - IIT Delhi\CPCB\PROF. HARI\data_final\CLASSIFICATION\df_cat_pro.csv')
#Delete the unwanted column
data_cat_pro = data_cat_pro.drop(['Unnamed: 0'], axis = 1)
#Delete the duplicate entries
data_cat_pro = data_cat_pro.drop_duplicates(['sentence'],keep = 'first').reset_index(drop = True)  
#Shows the first 5 rows
data_cat_pro.head()
```

```python
#Cleaning the each sentence in data frame
sentence = []
import re
for i in np.arange(0,data_cat_pro.shape[0], 1):
    pattern = r'\[[^()]*\]'
    t = re.sub(pattern,'', data_cat_pro.sentence[i])
    sentence.append(t)
data_cat_pro['sentence'] = sentence
#Shows first 5 rows
data_cat_pro.head()
```

```python
#String to be searched
string = 'catalyst'
add2 = []
#Custom dictionary of process parameters
process = ['temperature', 'pressure', 'weight', 'concentration', 'molar ratio',
           'molarity','Molality','gram','atm','celcius', 'kelvin' , 'volume', 'porocity']
replacements = {'K': 'kelvin', 'atm': 'pressure',
               '°C': 'celcius','bar': 'pressure',
                'g' : 'gram','m': 'molarity',
                'M': 'Molality','Mpa': 'pressure',
                'pa' : 'pressure', }
#Help to deal the units
replacer = replacements.get

for i in np.arange(0,data_cat_pro.shape[0],1):
    s = data_cat_pro.sentence[i]
    sentence_list = word_tokenize(s)
    sentence_list = [replacer(n, n) for n in sentence_list]
    s = " ".join(sentence_list)
    for pro in range(len(process)):
        if string in s:
            if process[pro] in s:
                #Extract chemical species using chemdataextractor
                doc = Document(s)
                ent = doc.cems                       
                comp = []
                for i in range(len(ent)):
                    #Convert span object to str
                    comp.append(ent[i].text)           
                if ent != []:
                    #Sentiment analysis
                    si = flair.data.Sentence(s)         
                    flair_sentiment.predict(si)
                    total_sentiment = si.labels
                    d = total_sentiment[0].to_dict()
                    if d['value'] == 'POSITIVE':
                        comp = set(comp)
                        if s[-1] == '.':
                            s = s[:-1]
                            sl = word_tokenize(s)
                            x  = word_tokenize(s)
                            add2.append(['__','__','__','__'])
                            for i in range(len(sl)):
                                if sl[i] in process:
                                    sl[i] = 'process'
                                elif sl[i] in name_catalyst:
                                    sl[i] = 'catalyst'
                                else:
                                    sl[i] = 0
                                add2.append([x[i],'NN', 0, sl[i]])   
```

```python
#Stored in data frame 
df = pd.DataFrame(add2)
#Delete the first row
df =  df.iloc[1: , :]
#Reset the index and drop previous
df = df.reset_index(drop = True)
#Shows the first 5 rows
df.head()
```
